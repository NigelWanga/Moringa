```python
%load_ext sql

# Then connect to our in memory sqlite database
# NB: This database will cease to exist as soon as the database connection is closed. 
%sql sqlite://
```

```python
import pandas as pd

with open('clean_cars.csv','r') as f:
    cars = pd.read_csv(f, index_col=0, encoding='utf-8')
%sql DROP TABLE if EXISTS cars;
%sql PERSIST cars;
%sql SELECT * FROM cars LIMIT 5;
```

```python
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
#
%%sql
SELECT City, hour, COUNT(hour) FROM cars
WHERE City = 'Paris' AND month = 4
GROUP BY hour
ORDER BY COUNT(hour) DESC LIMIT 1;
```

```python
#What is the most popular hour for returning cars?
#
%%sql
SELECT hour, COUNT(hour) FROM cars
WHERE Slots > 0
GROUP BY hour
ORDER BY COUNT(hour) DESC LIMIT 1;
```

```python
#What station is the most popular?
#Overall?
#
%%sql
SELECT "Station type", COUNT("Station type") FROM cars
GROUP BY "Statioin type"
ORDER BY  COUNT("Station type")
```

```python
#popular station that is most popular at picking hour
#
%%sql
SELECT "Station type", COUNT("Station type") FROM cars
WHERE hour =  21
GROUP BY "Statioin type"
ORDER BY  COUNT("Station type")
```

```python
#What postal code is the most popular for picking up Blue cars
#
%%sql
SELECT "Postal code", COUNT("Postal code") FROM cars
WHERE "Bluecar counter" > 0
GROUP BY "Postal code"
ORDER BY COUNT("Postal code") DESC LIMIT 1;
```

```python
#Do the results change if you consider Utilib and Utilib 1.4 instead of Blue cars
#
%%sql
SELECT "Postal code", COUNT("Postal code") FROM cars
WHERE "Utilib counter" > 0
GROUP BY "Postal code"
ORDER BY COUNT("Postal code") DESC LIMIT 1;
```

```python
%%sql
SELECT "Postal code", COUNT("Postal code") FROM cars
WHERE "Utilib 1.4 counter" > 0
GROUP BY "Postal code"
ORDER BY COUNT("Postal code") DESC LIMIT 1;
```
